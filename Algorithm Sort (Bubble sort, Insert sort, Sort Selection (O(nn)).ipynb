```python
a =[1, 2,43, 12,56, 21,0,5]
print(len(a))

# Sắp xếp nổi bột (bubble sort), làm nổi bột các phần tử lớn nhất cuối cùng ( O(n**2)): không nên dùng sắp xếp này vì độ phức tạp lớn
# so sánh 2 phần tử liền kề nhau
def bubble_sort(a):
    n = len(a)
    for i in range(n-1):
        print(i)
        for j in range(n-1 -i): # vì các phần tử cuối cùng đã được sắp xếp nên bỏ qua
            if(a[j]> a[j+1]):
                a[j] , a[j+1] = a[j+1], a[j]
            print('Bubble Sort :','i', i, 'sort: ', a)
bubble_sort([1, 2,43, 12,56, 21,0,5])
# Sắp xếp chèn ( Insert Sort): so sánh phần tử thứ i (key) với các phần tử trong mảng đã được sắp xếp (O(n**2)): không nên dùng
def insert_sort(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i-1
        while j >=0 and key< a[j]: # trong Python : và dùng and ( not &&)
            a[j+1] = a[j]
            j -=1
        a[j + 1] = key
        print('\n Insert_sort: ', 'i',i,'a: ',a)
insert_sort(a)
# Sắp xếp hay dùng: so sánh phần tử i với tất cả phần tử còn lại ( O(n**2))
def sort_selection (a):
    n = len(a)
    for i in range(n-1):
        for j in range(i+1,n):
            if(a[i]> a[j]):
                a[i], a[j] = a[j], a[i]
            print('i',i,'sort selection: ', a)
        print('\n')
sort_selection(a)
```
